# Let's describe each column for better understanding:

Win(W ) : number of games where pitcher was pitching while their team took the lead and went on to win, also the starter needs to pitch at least 5 innings of work

Runs(R):  times reached home plate legally and safely

At Bats(AB): plate appearances, not including bases on balls, being hit by pitch, sacrifices, interference, or obstruction

Hits(H): reaching base because of a batted, fair ball without error by the defense

Doubles(2B): hits on which the batter reaches second base safely without the contribution of a fielding error

Triples(3B): hits on which the batter reaches third base safely without the contribution of a fielding error

Homeruns(HR): hits on which the batter successfully touched all four bases, without the contribution of a fielding error

Walks(BB): hitter not swinging at four pitches called out of the strike zone and awarded first base.

Strikeouts(SO): number of times that a third strike is taken or swung at and missed, or bunted foul. Catcher must catch the                    third strike or batter may attempt to run to first base.


Stolen Bases(SB):number of bases advanced by the runner while the ball is in the possession of the defense

Runs Allowed(RA): The amount of runs scored against a pitcher. This includes earned runs and unearned runs.

Earned Runs(ER): number of runs that did not occur as a result of errors or passed balls

Earned Run Average (ERA):  total number of earned runs (see "ER" above), multiplied by 9, divided by innings pitched

Shutouts(SHO): number of complete games pitched with no runs allowed

Saves(SV): number of games where the pitcher enters a game led by the pitcher's team, finishes the game without surrendering the lead, is not the winning pitcher, and either (a) the lead was three runs or fewer when the pitcher entered the game; (b) the potential tying run was on base, at bat, or on deck; or (c) the pitcher pitched three or more innings.

Complete Games(CG): A pitcher earns a complete game if he pitches the entire game for his team regardless of how long it lasts.

Errors(E) :  number of times a fielder fails to make a play he should have made with common effort, and the offense benefits as a result

# Scaling the Data

```python
#Let's split the data into features and label
X=data.drop('W', axis=1)
y=data.W
```

```python
from sklearn.preprocessing import StandardScaler
scalar= StandardScaler()
X_scaled=scalar.fit_transform(X)
```

```python
X_scaled.shape[1]     # here [1] is index 1, which shows the number of columns
```

# VIF


Let's check if any multicolinearity is present in the data.

```python
#finding variance inflation factor in each scaled column i.e X_scaled.shape[1] (1/1-R2): here R2=actual-predicted
#this method help us to identify multicollinearity.

# Import library for VIF
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif=pd.DataFrame()
vif["vif"]= [variance_inflation_factor(X_scaled,i) for i in range(X_scaled.shape[1])]
vif["Features"]=X.columns

#let's check the values
vif
```

All the vif values are under 5, which means our data has no multicolinearity , So we can Proceed with our Model Making with the scaled data.

```python
# let's chcek the dimension of our scaled feature and label
X_scaled.shape
```

```python
y.shape
```

The data looks fine, We are good to proceed further with model building.


# Model Building


```python
#importing the library to split the data into train and test data.

from sklearn.model_selection import train_test_split, GridSearchCV

# spliting the data, in train and test dataset, we will be providing test size of 25% and train of 80%
X_train,X_test,y_train,y_test= train_test_split(X_scaled, y, test_size=0.25, random_state= 42)
```

```python
# importing all metrics

from sklearn.metrics import mean_absolute_error as MAE
from sklearn.metrics import mean_squared_error as MSE
from sklearn.metrics import r2_score as R2
```

```python
# importing ML Models

from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.linear_model import Lasso
import xgboost as xgb
xgb= xgb.XGBRegressor()
```

# Linear Regression Model

```python
# Model
LR= LinearRegression(normalize=True)

#fit 
LR.fit(X_train,y_train)

#predict
y_pred= LR.predict(X_test)

#score variables
LR_MAE= round(MAE(y_test, y_pred), 2)
LR_MSE= round(MSE(y_test, y_pred), 2)
LR_R2= round(R2(y_test, y_pred), 4)
LR_RMSE=round(np.sqrt(MAE(y_test,y_pred)),4)


print(f" Mean Absolute Error : {LR_MAE}\n")
print(f" Mean Squared Error : {LR_MSE}\n")
print(f" R^2 Score : {LR_R2}\n")
print(f" Root Mean Squared Error : {LR_RMSE}\n")
```

```python
## Cross validation of the models using K fold

from sklearn.model_selection import KFold 
 
kf = KFold(n_splits=3, random_state=1, shuffle=True)
scores = []

for i, (train_index, test_index) in enumerate(kf.split(X)):
    x_train = X_scaled[train_index,:]
    Y_train = y[train_index]
    x_test = X_scaled[test_index,:]
    Y_test = y[test_index]

    LR.fit(x_train, Y_train)
    test_score = LR.score(x_test, Y_test)
    scores.append(round(test_score, 4))

print("Scores for Linear Regression : ",scores)
average_LR= (scores[0]+scores[1]+scores[2])/3
print("Mean score is : ",average_LR)
```

# RandomForestRegressor

```python
# model
RFR= RandomForestRegressor()
#fit 
RFR.fit(X_train,y_train)

#predict
y_pred= RFR.predict(X_test)

#score variables
RFR_MAE= round(MAE(y_test, y_pred), 2)
RFR_MSE= round(MSE(y_test, y_pred), 2)
RFR_R2= round(R2(y_test, y_pred), 4)
RFR_RMSE=round(np.sqrt(MAE(y_test,y_pred)),4)

print(f" Mean Absolute Error : {RFR_MAE}\n")
print(f" Mean Squared Error : {RFR_MSE}\n")
print(f" R^2 Score : {RFR_R2}\n")
print(f" Root Mean Squared Error : {RFR_RMSE}\n")
```

```python
# Cross validation of the models using K fold

from sklearn.model_selection import KFold 
 
kf = KFold(n_splits=3, random_state=1, shuffle=True)
scores = []

for i, (train_index, test_index) in enumerate(kf.split(X)):
    x_train = X_scaled[train_index,:]
    Y_train = y[train_index]
    x_test = X_scaled[test_index,:]
    Y_test = y[test_index]

    RFR.fit(x_train, Y_train)
    test_score = RFR.score(x_test, Y_test)
    scores.append(round(test_score, 4))

print("Scores for RandomForestRegressor : ",scores)
average_RFR=(scores[0]+scores[1]+scores[2])/3
print("Mean score for RFregressor : ", average_RFR)
```

```python
# Hyper parameter tuning

rfr= RandomForestRegressor()

#determining what algorithm and parameter works better.
param_grid= {'n_estimators': [40,100,200],
            'criterion': ['squared_error', 'absolute_error'],
             'max_depth': [2,10,15],
             'min_samples_leaf':[2,5,8],
            'n_jobs': [1,4,6]}

grid_search=GridSearchCV(estimator=rfr, param_grid= param_grid, cv=5)

grid_search.fit(X_train,y_train)

grid_search.best_estimator_
```

```python
grid_search.best_params_
```

```python
# we will use these best parameters in Random forest algorithm and check if accuracy is increasing.

rfr=RandomForestRegressor(criterion='squared_error', max_depth=10,
                      min_samples_leaf=2, n_estimators=200, n_jobs=6)
rfr.fit(X_train,y_train)
y_predi=rfr.predict(X_test)   #predicting with best parameter

print("Accuracy post tuning \n")
print(R2(y_test,y_predi))
```
