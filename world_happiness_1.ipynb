We will drop Country Region and Happiness rank from the model since these features are not important in model building

```python
data=data.drop(['Country','Region','Happiness Rank'], axis=1)
```

```python
data.head(2)
```

```python
data.dtypes
```

```python
X= data[['Economy (GDP per Capita)', 'Family',
       'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)',
       'Generosity', 'Dystopia Residual']]
y=data['Happiness Score']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
lr=LinearRegression()
```

```python
X_train,X_test,y_train,y_test= train_test_split(X,y,random_state=40,test_size=0.3)
lr.fit(X_train,y_train)
lr.score(X_test,y_test)
   
```

```python
y_pred=lr.predict(X_test)
```

```python
print(y_pred)
```

```python
print(y_test)
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.plot(y_test,y_pred)
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor_forest = RandomForestRegressor(n_estimators = 8, random_state = 42)
regressor_forest.fit(X_train, y_train)
regressor_forest.score(X_test,y_test)
```
